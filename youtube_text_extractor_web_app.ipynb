```python
%%writefile app.py

from __future__ import unicode_literals
import streamlit as st
import yt_dlp as youtube_dl


def main():
    ydl_opts = {'outtmpl': '/content/files/download.%(ext)s',
    'format': '(bestvideo[width>=1080][ext=mp4]/bestvideo)+bestaudio/best', #Ensures best settings
    'writesubtitles': False, #Adds a subtitles file if it exists
    'writeautomaticsub': False, #Adds auto-generated subtitles file
    'subtitle': '--write-sub --sub-lang en', #writes subtitles file in english
    'subtitlesformat':'srt', #writes the subtitles file in "srt" or "ass/srt/best"
    'skip_download': False, #skips downloading the video file
    }
    url = st.text_input("Please add YouTube link here:") #When this appears you insert your YouTube Video Link
    if url!='':
              with youtube_dl.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
              print("Download Successful!")


              from pytube import YouTube
              from PIL import Image
              import requests
              from io import BytesIO

              # URL of the YouTube video you want to download
              video_url = url

              # Create a YouTube object
              yt = YouTube(video_url)

              # Get the URL of the video's thumbnail
              thumbnail_url = yt.thumbnail_url

              # Download the thumbnail image
              response = requests.get(thumbnail_url)
              thumbnail_image = Image.open(BytesIO(response.content))

              # Save the thumbnail image
              thumbnail_save_path = "/content/files/thumbnail.jpg"
              thumbnail_image.save(thumbnail_save_path)

              st.image("/content/files/thumbnail.jpg")


              from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_audio
              import os

              # Path to the downloaded .mp4 video file
              video_path = "/content/files/download.webm"

              # Output path for the converted .mp3 audio file
              audio_path = "/content/files/converted_audio.wav"

              # Convert the video to audio
              ffmpeg_extract_audio(video_path, audio_path)

              print("Conversion completed: .mp4 to .wav")

              import speech_recognition as sr

              audio_path = "/content/files/converted_audio.wav"

              # Initialize the recognizer
              recognizer = sr.Recognizer()

              # Load the audio file
              with sr.AudioFile(audio_path) as source:
                  audio = recognizer.record(source)

              # Perform speech recognition
              try:
                  text = recognizer.recognize_google(audio)
                  print("Transcription:")
                  print(text)
              except sr.UnknownValueError:
                  print("Speech recognition could not understand audio")
              except sr.RequestError as e:
                  print("Could not request results from Google Speech Recognition service; {0}".format(e))
              from google.colab import files
              code = text
              with open('/content/files/transcription.txt', 'w') as f:
                  f.write(code)
              st.write (text)

              import shutil
              shutil.make_archive("output", 'zip', "/content/files")


              with open("/content/output.zip", "rb") as fp:
                  btn = st.download_button(
                    label="Download ZIP",
                    data=fp,
                    file_name="output.zip",

                    )

              with open("/content/files/transcription.txt", "rb") as fp:
                  btn = st.download_button(
                    label="Download transcription file",
                    data=fp,
                    file_name="transcription.txt",

                    )
    else:
        st.write("Rerun the code, invaild url")

if __name__ == '__main__':
    main()
    import shutil
    shutil.rmtree("/content/files")
```
